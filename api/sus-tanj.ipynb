```python
from requests import api
```

```python
r = api.get('http://127.0.0.1:8000/audio/list')
r
```

```python
r = api.post('http://127.0.0.1:8000/audio/create', json={
    "text": "sus tanju",
    "voice": 'tr-TR-AhmetNeural',
    "rate": 0,
    "volume": 0,
    "pitch": 0,
    },
    stream=True
)
```

```python
r.raise_for_status()

with open("audio.mp3", "wb") as f:
    for chunk in r.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
```
