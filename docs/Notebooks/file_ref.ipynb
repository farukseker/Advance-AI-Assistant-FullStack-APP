```python
from pathlib import Path
from slugify import slugify
```

```python
BASE_DIR = Path().resolve()
```

```python
test_doc_dir = BASE_DIR / 'The Django admin site _ Django documentation _ Django.pdf'
```

```python
slugify(test_doc_dir.stem) + test_doc_dir.suffix
```

---

```python
from requests import api
url = "http://127.0.0.1:8000/embed/ask-with-file"
```

```python

filename = slugify(test_doc_dir.stem) + test_doc_dir.suffix

files = {
    "file": (
        filename,
        test_doc_dir.read_bytes(),
        "application/pdf" if test_doc_dir.suffix == ".pdf" else "text/plain"
    )
}
```

```python

data = {
    "question": "tell me how it works django admin sites answer as turkish",
    "top_k": "3"
}

response = api.post(url, files=files, data=data)
    
result = response.json()
print(result)
```
