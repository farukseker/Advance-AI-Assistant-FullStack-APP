```python
big_text = """

Giriş:
Bu videoda 2026 için öğrenilmesi en faydalı programlama dilleri sıralanıyor. Önemli olan trendler değil. Gerçek üretimde hangi diller kullanılıyor, hangi rollerde talep var ve AI hangi becerileri güçlendiriyor — bunlar dikkate alındı.

1. TypeScript:
TypeScript, modern uygulamaların varsayılan dili oldu. GitHub verilerine göre en çok katkı alan dillerden biri. Artık sadece ön yüz değil; arka uç, iç araçlar, geliştirici araçları ve AI ürünleri için de kullanılıyor. Birçok startup TypeScript’i uçtan uca kullanıyor. AI arayüz kodu üretebilir ama durum yönetimi, veri modelleri ve mimari kararları hâlâ insan ekspertizine ihtiyaç duyuyor. Gerçek ürünler geliştirmek istiyorsanız TypeScript yüksek verim sağlar.

2. Python:
Python, geliştiriciler arasında yaygın ve büyüyen bir dil. Hızı için değil; sistemleri orkestra etmesi, veri akışlarını yönetmesi, iş otomasyonu ve AI boru hatlarını kontrol etmesi için güçlü. Ağır hesaplamalar C++ veya Rust’ta olsa bile Python genelde üstte koordinasyonu yapar. Ancak Python artık daha kolay yazılabiliyor, bu yüzden tek başına nadir bir yetenek değil. En iyi strateji: Python’u başka bir uzmanlıkla birleştirmek. Örneğin Python + AI, Python + altyapı veya domain bilgisi.

3. Go:
Go, bulut altyapısı, DevOps araçları ve yüksek verimli arka uç servislerinde yaygın. Go işleri genelde daha az başvuru alır ama beklentiler yüksektir; çünkü Go geliştiricileri genelde sistemleri ve üretimi iyi anlar. Go rolleri sorumluluk ve güven gerektirir; bu da maaşları ve kariyer fırsatlarını yükseltir. Altyapı ve yüksek ölçekli servislerle ilgileniyorsanız Go iyi bir yatırım.

4. Rust:
Rust, geliştiriciler tarafından çok beğeniliyor ama hâlâ yaygınlık kazanıyor. Sistem programlama, güvenlik hassas yazılımlar, blockchain ve performans kritik uygulamalarda tercih ediliyor. Rust, bellek hatalarını ve güvenlik açıklarını önlemekte çok başarılı. AI’nın sahte yetkinlik üretmesinin zor olduğu dillerden biri; yani Rust bilenler daha zor ikame edilir. Donanım seviyesine yakın ve uzun ömürlü yazılımlar yapmak istiyorsanız Rust sağlam bir tercih.

5. Java ve C#:
Bu iki dili benzer kategoriye alıyoruz. Kurumsal sistemlerde, finansal platformlarda ve büyük, eski altyapılarda hâlâ yaygın olarak kullanılıyorlar. Bankalar, büyük şirketler ve hükümet sistemleri bu dillerle çalışıyor. Eğer stabil, kurumsal roller ve karmaşık gerçek dünya sistemleri istiyorsanız Java veya C# bilmek avantaj sağlar. Özellikle başlangıç seviyesinde bu dillerle öne çıkmak mümkün.

Kapanış ve tavsiye:
En büyük hata diller toplamak. Bunun yerine bir ana dil seçin ve derinleşin. Sonra bir “çarpan” (örneğin altyapı, AI veya domain uzmanlığı) ekleyin. 2026’da derinlik, yüzeyselliğe göre daha değerli olacak. Hangi dili öğreneceğinizi yorumlarda belirtin.

""".strip()
```

```python
r = api.post('http://127.0.0.1:8000/audio/create', json={
    "text": big_text,
    "voice": 'tr-TR-AhmetNeural',
    "rate": 0,
    "volume": 0,
    "pitch": 5,
    },
    stream=True
)
```

```python
r.raise_for_status()

with open("audio3.wav", "wb") as f:
    for chunk in r.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
```
