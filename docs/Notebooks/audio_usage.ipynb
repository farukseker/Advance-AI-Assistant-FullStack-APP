```python
big_text = """
10.000’den fazla geliştiricinin teknik mülakatlara hazırlanmasına yardımcı oldum.
""".strip()
```

```python
r = api.post('http://127.0.0.1:8000/audio/create', json={
    "text": big_text,
    "voice": 'tr-TR-AhmetNeural',
    "rate": 0,
    "volume": 0,
    "pitch": 5,
    },
    stream=True
)
```

```python
r.raise_for_status()

with open("audio2.wav", "wb") as f:
    for chunk in r.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
```
