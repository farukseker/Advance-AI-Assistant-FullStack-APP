```python
available_tts_list = []
```

```python
audio_list = api.get('http://127.0.0.1:8000/audio/list')
```

```python
for audio in audio_list.json():
    if audio.get('Locale') in ['tr-TR', 'en-US', 'az-AZ']:
        available_tts_list.append(audio.get('ShortName'))
        
```
