```python
from requests import api
```

```python
with open('The Django admin site _ Django documentation _ Django.pdf', 'rb') as f:
    files = {'file': f} 
    r = api.post('http://localhost:8000/embed/load-document', files=files)
    
print(r.status_code)
print(r.text)
```
