```python
import requests
import json

url = f"http://127.0.0.1:8000/ai/chat/{chat_id}"
def ask(_ask: str, prompt: str = ''):
    with requests.post(
        url,
        json={
            "question": _ask,
        },
        stream=True,
        timeout=60
    ) as r:
        r.raise_for_status()
    
        for line in r.iter_lines(decode_unicode=True):
            if not line:
                continue
    
            if line.startswith("data:"):
                chunk = line.removeprefix("data:").strip()
    
                if chunk == "[DONE]":
                    break
                
                ch = json.loads(chunk) 
                if ch.get('type') == 'tool_start':
                    print('Start:', ch.get('tool'), 'q: ', ch.get('input'))
                    print('\n')
                
                if ch.get('type') == 'tool_end':
                    print('End:', ch.get('tool'))
                    print('\n')
                
                print(ch.get('content'), end="", flush=True)
```

```python
ask(
    "bu video daki bilgiyi bana türkçe açıkla açıklaman bir tts modeline verilip seslendirilecek buna dikkat ederek yazarsan sevinirim https://www.youtube.com/watch?v=zsPD9aohymw"
)
```
